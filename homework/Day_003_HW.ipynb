## [作業]
Real-Time Recognition of Handwritten Chinese Characters Spanning a Large Inventory of 30,000 Characters


1. 專案目標

基於 deep learning 技術，辨識手寫漢字，並使用在平板、智慧型手機等可手寫輸入的平台。

2. 使用技術

使用 CNN 做手寫辨識。


Input為48x48的漢字手寫圖像，以類似 <CASIA Online and Offline Chinese Handwriting Databases, in Proc. 11th Int. Conf. Document Analysis Recognition (ICDAR 2011), Beijing, China, Sept. 2011.> 的方法分成 training set 和 testing set， 其中有一百萬個 training set. 

3. 資料來源
 https://machinelearning.apple.com/2017/09/12/handwriting.html
