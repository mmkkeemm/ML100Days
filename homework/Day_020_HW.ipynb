```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
import random as rn
matrix1=[]
for i in range(0,10):
    list1=[]
    for j in range(0,10):
        list1.append(rn.uniform(-1.0,1.0))
    matrix1.append(list1)
matrix=pd.DataFrame(matrix1)
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix,cmap = plt.cm.RdYlBu_r)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
import random as rn
matrix2=[]
for i in range(nrow):
    list2=[]
    for j in range(ncol):
        list2.append(rn.uniform(-1.0,1.0))
    matrix2.append(list2)
matrix=matrix2
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()
list(plot_data.columns)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot,cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix3=[]
for i in range(nrow):
    list3=[]
    for j in range(ncol):
        list3.append(np.random.randn())
    matrix3.append(list3)
matrix=matrix3

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot,cmap = plt.cm.OrRd_r)

plt.show()
```
